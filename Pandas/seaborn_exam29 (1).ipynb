```python
#1

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
x,y=make_classification(n_samples=1000,n_features=10,weights=[0.9,0.1],random_state=42)
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=LogisticRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
cm=confusion_matrix(y_test,y_pred)
sns.heatmap(cm,annot=True,fmt='d',cmap="coolwarm")
plt.title("confusion matrix")
plt.xlabel('predicted')
plt.ylabel('true')
plt.show()
```

```python
#12
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
subjects = ['Python', 'English', 'Robotics']
batch_a = [60, 95, 65]
batch_b = [70, 85, 75]
batch_c = [80, 90, 85]
fig, ax = plt.subplots(figsize=(10, 6))
x = np.arange(len(subjects))
width = 0.25
ax.bar(x - width, batch_a, width, label='Batch A', color='red')
ax.bar(x, batch_b, width, label='Batch B', color='green')
ax.bar(x + width, batch_c, width, label='Batch C', color='black')
ax.set_title('Performance of Each Batch Across Different Subjects')
ax.set_xlabel('Subjects')
ax.set_ylabel('Performance')
ax.set_xticks(x)
ax.set_xticklabels(subjects)
ax.legend()
plt.show()
#These trends and patterns provide insights into the performance of each batch across different subjects, highlighting areas of strength and weaknessIn Python, Batch A (red) performed the highest, followed closely by Batch C (black). Batch B (green) has the lowest performance.
#Batch C outperforms other batches: Across all three subjects (Python, English, and Robotics), Batch C consistently scores higher than Batch A and Batch B.
# Batch A struggles with Python: Batch A scores significantly lower in Python compared to the other two batches.
# Batch B shows consistent performance: Batch B's scores are relatively consistent across all three subjects, with no significant outliers.
# English is the strongest subject: Across all three batches, English appears to be the strongest subject, with the highest scores.
# Robotics scores are closer together: Compared to Python and English, the scores for Robotics are closer together across the three batches, indicating less variation.
```

```python
#14
import seaborn as sns
import matplotlib.pyplot as plt

diamonds = sns.load_dataset("diamonds")
plt.figure(figsize=(8, 6))
sns.scatterplot(data=diamonds, x="carat", y="price", hue="cut")
plt.title("Relationship Between Price and Carat")
plt.show()
plt.figure(figsize=(8, 6))
sns.boxplot(data=diamonds, x="cut", y="price")
plt.title("Relationship Between Price and Cut")
plt.show()
plt.figure(figsize=(8, 6))
sns.scatterplot(data=diamonds, x="depth", y="price", hue="cut")
plt.title("Relationship Between Price and Depth")
plt.show()
plt.show()
#Key Findings

# Carat: Carat weight is the most significant driver of diamond prices. Larger diamonds are generally priced higher.
# Cut: Diamond cut quality significantly influences price. Ideal and Premium cuts command higher prices due to their superior brilliance and sparkle.
# Depth and Table: Depth and table measurements have a less pronounced impact on price compared to carat and cut.
# Clarity and Color: These features are implicitly considered through the cut quality. While not directly visualized, higher clarity and better color grades contribute to a higher price.
```
